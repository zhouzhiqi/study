# Seaborn

基于matplotlib，用更简单的调用呈现更多美图

```python
# 工具包
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.__version__ # '0.7.1' dont need sns.set()
```

# 与plt对比

```python
np.random.seed(sum(map(ord, "aesthetics"))) # 给每一个任务单独分配随机种子
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7): # up to but not include
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
```

```python
sns.set()
sinplot()
```

# 风格参数

## 画布背景

darkgrid/whitegrid（深色带格子／白色带格子）
dark/white：更关注整体趋势
ticks：强化刻度
网格有助于定量分析数据

```python
style = ["darkgrid", "whitegrid", "dark", "white", "ticks"]
# 默认darkgrid
for s in style:
    sns.set_style(s)
    plt.title(s)
    sinplot()
    plt.show()
```

## 坐标轴参数

```python
sinplot()
sns.despine() # despine()函数是去掉了上右边框
# sns.despine(fig=None, ax=None, top=True
#             , right=True, left=False, bottom=False
#             , offset=None, trim=False)
```

```python
sinplot()
sns.despine(offset=50, trim=True) 
# offset参数可以调节分离坐标轴
# trim调整坐标轴刻度位置
```

## 应用背景

```python
context = ["paper", "notebook", "talk", "poster"] 
# 默认notebook
for c in context:
    sns.set_context(c)
    # sns.set_context(context=None, font_scale=1, rc=None)
    plt.title(c)
    sinplot()
    plt.show()
```

## 调色板

### 颜色主题

```python
theme = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]
for t in theme:
    current_palette = sns.color_palette(t)
    # sns.color_palette(palette=None, n_colors=None, desat=None)
    sns.palplot(current_palette)
    plt.title(t)
    plt.show()
```

### 分类调色板
适用于分类描述的数据，而且分类之间无序

    hls颜色系统（默认），色调+亮度+饱和度
       （hue + luminance + saturation）
    husl颜色系统，hls中不同颜色存在视差，
       eg.黄色偏亮，蓝色偏暗，husl使得更加平衡

```python
# 恢复默认设置
sns.set()
# 绘制当前调色板
current_palette = sns.color_palette()
sns.palplot(current_palette)
# 绘制hls系统下的10色调色板（黄色偏暗 蓝色偏亮）
sns.set_palette(sns.color_palette("hls", 10))
current_palette = sns.color_palette()
sns.palplot(current_palette)
# 绘制husl系统下的10色调色板（与hls相比，色彩不突兀）
sns.set_palette(sns.color_palette("husl", 10)) # http://www.hsluv.org
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

### 序列调色板
主要通过亮度和饱和度区分有序分类
关注单级数据

```python
sns.palplot(sns.color_palette("Greens",10))
```

```python
with sns.color_palette("Greens_d"):
    sinplot()
```

### 分叉调色板¶
关注两级数据

```python
sns.palplot(sns.color_palette("autumn",11)) # 比如表示温度梯度
sns.palplot(sns.diverging_palette(10, 220, sep=80, n=11))
```

```python
sns.palplot(sns.diverging_palette(255, 133, l=60
                                  , n=10, center="dark"))
```

```python
with sns.diverging_palette(255, 133, l=60, n=5, center="dark"):
    sinplot()
```

### 自定义调色

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
sns.palplot(sns.color_palette(flatui))
```
