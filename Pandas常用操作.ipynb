```python
train_corr = train.corr().abs() #相关性矩阵
train_corr_index = train_corr.index
size = train_corr.shape[1] #for 的范围
threshold = 0.6 #相关性分界点

corr_list = [] 
# 存储格式: [相关性大小, 轴i, 轴j]
for i in range(0, size):
    for j in range(i+1, size):
        corr_list.append([train_corr.iloc[i, j], i, j])  

# 按相关性大小进行排序
corr_list_sorted = sorted(corr_list, key=lambda x:-x[0])

# 输出相关性大于threshold的
# [相关性大小, 轴i, 轴j]
for v, i, j in corr_list_sorted:
    if v > threshold:
        print ("%.3f is %s and %s" % (v
                                      , train_corr_index[i]
                                      , train_corr_index[j]))
```
